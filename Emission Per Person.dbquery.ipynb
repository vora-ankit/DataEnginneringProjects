```sql
SELECT 
  county_state_name,
  population,
  CAST(
    REGEXP_REPLACE(`GHG emissions mtons CO2e`, '[^0-9.]', '') AS DOUBLE
  ) / CAST(
    REGEXP_REPLACE(population, '[^0-9.]', '') AS DOUBLE
  ) AS emissions_per_person
FROM 
  emissions.default.emissions_data AS ed
  ORDER BY emissions_per_person DESC
LIMIT 
  10
```
